```python
SnowballTarget.yaml

behaviors:
  SnowballTarget:
    trainer_type: ppo
    summary_freq: 50000
    keep_checkpoints: 10
    checkpoint_interval: 500000
    max_steps: 2000000
    time_horizon: 256
    threaded: true
    hyperparameters:
      learning_rate: 0.0003
      learning_rate_schedule: linear
      batch_size: 128
      buffer_size: 10240
      beta: 0.005
      epsilon: 0.2
      lambd: 0.95
      num_epoch: 3
    network_settings:
      normalize: false
      hidden_units: 256
      num_layers: 3
      vis_encode_type: simple
    reward_signals:
      extrinsic:
        gamma: 0.99
        strength: 1.0
```
